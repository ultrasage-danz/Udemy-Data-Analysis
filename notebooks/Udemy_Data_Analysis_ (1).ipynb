> Top selling courses (and visualisation)

```python
df.sort_values('num_subscribers', ascending=False)
```

```python
#do a chart for the top selling courses
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by 'num_subscribers' in descending order and select the top 10
top_courses = df.sort_values('num_subscribers', ascending=False).head(10)

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='num_subscribers', y='course_title', data=top_courses, palette='viridis')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Title')
plt.title('Top 10 Selling Courses')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

> All courses related with Python

```python
df[df['course_title'].str.contains('Python')]
```

```python
import matplotlib.pyplot as plt

# Count Python courses
python_courses = df[df['course_title'].str.contains('Python', case=False)].shape[0]

# Count total courses
total_courses = df.shape[0]

# Calculate percentages
python_percent = (python_courses / total_courses) * 100
other_percent = 100 - python_percent

# Create the pie chart
labels = ['Python Courses', 'Other Courses']
sizes = [python_percent, other_percent]
colors = ['#66b3ff', '#99ff99']

plt.pie(sizes, labels= labels, autopct='%1.1f%%', colors=colors, startangle=90,
        textprops={'fontsize': 12, 'weight': 'bold'})  # Increase font size and weight

plt.axis('equal')
plt.title('Percentage of Python Courses', pad=20)  # Add padding to the title


plt.show()
```

```python
# @title is_paid


from matplotlib import pyplot as plt
import seaborn as sns

# Calculate the counts for paid and free courses
course_counts = df['is_paid'].value_counts()

# Define complementary colors
colors = sns.color_palette("Paired", n_colors=2)

# Create the pie chart
plt.pie(course_counts, labels=course_counts.index, autopct='%1.1f%%', colors=colors)

# Add a title and legend
plt.title('Distribution of Free and Paid Courses')
plt.legend(labels=['Paid', 'Free'])

# Ensure the circle is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
```

```python
# @title level

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('level').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)

#add title and appropriate axes
plt.title('Number of Courses by Level')
plt.xlabel('Number of Courses')
plt.ylabel('Level')
```

```python
# @title subject

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('subject').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)

#add title and appropriate axes
plt.title('Number of Courses by Subject')
plt.xlabel('Number of Courses')
plt.ylabel('Subject')
```
